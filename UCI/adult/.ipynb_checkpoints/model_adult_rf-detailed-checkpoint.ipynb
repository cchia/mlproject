# Model on Adult Data from UCI Machine Learning Repository

The dataset is collected from the 1994 Census database, also known as "Census Income" dataset. 
The purpose of this project is to predict whether a person's income exceeds $50K annually, i.e., it is a binary classification problem. 

## Import Python Libraries

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

## Data Loading

The data is separated into two parts: training (adult.data) and validation (adult.test) datasets, with 14 features and 1 label (income), total 15 columns. 

```python
columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 
          'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 
          'hours-per-week', 'native-country', 'income']
```

```python
df = pd.read_csv('adult.data', names=columns)
```

```python
df_test = pd.read_csv('adult.test', names=columns, skiprows=1)
```

```python
def plotDist(column):
    temp_positive = df[df['income']==' >50K'].groupby(column)['income'].count().reset_index()
    temp_negative = df[df['income']==' <=50K'].groupby(column)['income'].count().reset_index()

    plt.figure(figsize=(20,10))
    plt.bar(temp_positive[column], temp_positive['income'], color='b')
    plt.bar(temp_negative[column], temp_negative['income'], color='r', alpha=0.5)    
```

### Remarks: 
1. There are missing values in 3 features: occupation, workclass and native-country, instead of dropping the missing values data, let's assign them with a new category -- unknown
2. Data is imbalanced in native-country, instead of treating each country as a separate category value, it may help to treat the data as binary -- US and non US. 
3. Same case applies to capital-gain and capital-loss, most data have 0 in capital-gain and capital-loss, instead of using original value, we can try binary value, 0 or >0.

## Missing value treatment

```python
df['type']='training'
df_test['type'] = 'test'
df_combine = pd.concat([df, df_test], sort=False)
```

```python
df_combine['occupation2'] = df_combine.apply(lambda row: 'unknown' if row['occupation']==' ?' else row['occupation'], axis=1)
df_combine['country'] = df_combine.apply(lambda row: 'unknown' if row['native-country']==' ?' else row['native-country'], axis=1)
df_combine['workclass2'] = df_combine.apply(lambda row: 'unknown' if row['workclass']==' ?' else row['workclass'], axis=1)
```

## Feature Engineering

Convert y to 0/1 value

```python
df_combine['y'] = df_combine.apply(lambda row: 1 if (row['income']==' >50K' or row['income']==' >50K.') else 0, axis=1)
```

Split the features into nominal and numerical parts

```python
nominal_cols = ['workclass2', 'education', 'marital-status', 'occupation2', 'relationship', 'race', 'sex', 'country']
numerical_cols = ['age', 'fnlwgt', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week']
```

### Treatment to nominal features
For nominal features in ML problems, usually there are 2 types of treatments: 
1. Ordinal (label) Encoding -- assign different integer values to different classes under the feature
2. One Hot (dummy) Encoding -- transform each category into a new feature with binary value
In this exercise, we can try both method and examine which method is better

## Label Encoding for Ordinal

```python
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
```

```python
oe = OrdinalEncoder()
scaler = StandardScaler()
```

Encode all nominal features

```python
transformed_features = oe.fit_transform(df_combine[nominal_cols].values)
```

```python
scaled_features = scaler.fit_transform(df_combine[numerical_cols].values)
```

```python
transformed_df = pd.DataFrame(transformed_features, index=df_combine.index, columns=df_combine[nominal_cols].columns)
```

```python
scaled_df = pd.DataFrame(scaled_features, index=df_combine.index, columns=df_combine[numerical_cols].columns)
```

Combine with numerical features after encoding

```python
df_merge = pd.concat([df_combine[numerical_cols+['type', 'y']], transformed_df], axis=1)
```

```python
df_merge2 = pd.concat([df_combine[['type', 'y']], transformed_df, scaled_df], axis=1)
```

## Finetune Model

Gradient Boosting Tree has the best performance overall, let's pick GBT and apply it to one hot encoded features and finetune its performance using different set of hyperparameters. To find out the best hyperparameters, let's use K fold cross validation and find out the parameters set that achieve highest score. 

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {'n_estimators': [100, 200], 'max_depth': [3, 4], 'learning_rate': [0.05, 0.01]}
```

```python
gbc3 = GridSearchCV(GradientBoostingClassifier(), params, cv=10, scoring='accuracy')
```

```python
gbc3.fit(x_train, y_train)
```

```python
gbc3.cv_results_
```

```python
gbc3.best_score_
```

```python
bestEstimator = gbc3.best_estimator_
```

```python
bestEstimator.score(x_test, y_test)
```

Compare to the score of test score without finetune, model after finetune improves the score from 0.867 to 0.874, slightly increases by 0.007
